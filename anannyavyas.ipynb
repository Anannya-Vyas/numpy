```python
def sum_of_subarray_ranges(arr):
    n = len(arr)

    # Helper function to find left and right bounds using a monotonic stack
    # compare_left_pop: condition to pop from stack for left_bound calculation
    # compare_right_pop: condition to pop from stack for right_bound calculation
    def get_bounds(arr, compare_left_pop, compare_right_pop):
        left_bound = [-1] * n
        right_bound = [n] * n

        stack = []
        for i in range(n):
            while stack and compare_left_pop(arr[stack[-1]], arr[i]):
                stack.pop()
            if stack:
                left_bound[i] = stack[-1]
            stack.append(i)

        stack = []
        for i in range(n - 1, -1, -1):
            while stack and compare_right_pop(arr[stack[-1]], arr[i]):
                stack.pop()
            if stack:
                right_bound[i] = stack[-1]
            stack.append(i)

        return left_bound, right_bound

    # Calculate sum of maximums for all subarrays
    # For arr[i] to be the maximum:
    #   left_max_idx: nearest index j < i such that arr[j] > arr[i] (strictly greater)
    #   right_max_idx: nearest index k > i such that arr[k] >= arr[i] (greater or equal)
    # The lambda functions define when to POP elements from the stack.
    # For left_max: pop if stack_val <= current_val (to find strictly greater left)
    # For right_max: pop if stack_val < current_val (to find greater or equal right)
    left_max, right_max = get_bounds(
        arr,
        lambda stack_val, current_val: stack_val <= current_val,
        lambda stack_val, current_val: stack_val < current_val
    )

    total_max_sum = 0
    for i in range(n):
        # Number of subarrays where arr[i] is the maximum
        count = (i - left_max[i]) * (right_max[i] - i)
        total_max_sum += arr[i] * count

    # Calculate sum of minimums for all subarrays
    # For arr[i] to be the minimum:
    #   left_min_idx: nearest index j < i such that arr[j] < arr[i] (strictly smaller)
    #   right_min_idx: nearest index k > i such that arr[k] <= arr[i] (smaller or equal)
    # For left_min: pop if stack_val >= current_val (to find strictly smaller left)
    # For right_min: pop if stack_val > current_val (to find smaller or equal right)
    left_min, right_min = get_bounds(
        arr,
        lambda stack_val, current_val: stack_val >= current_val,
        lambda stack_val, current_val: stack_val > current_val
    )

    total_min_sum = 0
    for i in range(n):
        # Number of subarrays where arr[i] is the minimum
        count = (i - left_min[i]) * (right_min[i] - i)
        total_min_sum += arr[i] * count

    return total_max_sum - total_min_sum
```

```python
# Test cases

# Example 1
arr1 = [1, 2, 3]
output1 = sum_of_subarray_ranges(arr1)
print(f"Input: {arr1}\nOutput: {output1}")
assert output1 == 4

# Example 2
arr2 = [-32, 0, -2, 72]
output2 = sum_of_subarray_ranges(arr2)
print(f"Input: {arr2}\nOutput: {output2}")
assert output2 == 318

# Additional test case
arr3 = [5, 2, 5]
# Subarrays: [5](0), [2](0), [5](0), [5,2](3), [2,5](3), [5,2,5](3)
# Sum of ranges: 0+0+0+3+3+3 = 9
output3 = sum_of_subarray_ranges(arr3)
print(f"Input: {arr3}\nOutput: {output3}")
assert output3 == 9

print("All test cases passed!")
```
