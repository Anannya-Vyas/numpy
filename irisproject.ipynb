# percentile
Below what value does P% of the data fall

```python
Unsorted:  [40, 10, 50, 20, 30,60]
Sorted:    [10, 20, 30, 40, 50,60]# percentiles are position-based, not value-based
# percentiles require ordered data
# decide which percentile you want
# find the position among data set with n values = p/100*(N-1)

```

```python
data = [40, 10, 50, 20, 30, 60]
sorted_data = sorted(data)
print(f"Unsorted: {data}")
print(f"Sorted: {sorted_data}")

# manual calculation of 50th percentile
# p/100 * (N-1)
#  50th percentile (p=50) and N=6 elements:
# position = 50/100 * (6-1) = 0.5 * 5 = 2.5
#means the 50th percentile is halfway between the element at index 2 and index 3 (0-indexed) in sorted data.
# sorted data: [10, 20, 30, 40, 50, 60]
#  index 2 is 30
#index 3 is 40
# 50th percentile = (30 + 40) / 2 = 35

percentile_50_manual = (sorted_data[2] + sorted_data[3]) / 2
print(f"Manually calculated 50th percentile: {percentile_50_manual}")
```

```python
import numpy as np

data = [40, 10, 50, 20, 30, 60]

#using numpy to find 50th percentileee
percentile_50_numpy = np.percentile(data, 50)
print(f"NumPy calculated 50th percentile: {percentile_50_numpy}")
```

doing wth iris


```python
import numpy as np

url = "https://raw.githubusercontent.com/sarwansingh/Python/refs/heads/master/ClassExamples/data/iris.csv"
npiris = np.loadtxt(url, delimiter=',', usecols=range(4))

sepal_length = npiris[:, 0]
print(sepal_length)
# mean sepal length
mean_sepal = np.mean(sepal_length)
print(mean_sepal)
print("Mean sepal length:", round(mean_sepal, 2))

```

```python
# percentiles
p25 = np.percentile(sepal_length, 25)
p50 = np.percentile(sepal_length, 50)# median
p75 = np.percentile(sepal_length, 75)

print("25th percentile:", p25)
print("50th percentile (median):", p50)
print("75th percentile:", p75)
```

```python
# percentage
total = npiris.shape[0]
# total elemets in that row of sepal leghth

print(total)
print('\n')

setosa = npiris[:50]
versicolor = npiris[50:100]
virginica = npiris[100:150]

print("Setosa %:", (setosa.shape[0] / total) * 100)
print('\n')
print("Versicolor %:", (versicolor.shape[0] / total) * 100)
print('\n')
print("Virginica %:", (virginica.shape[0] / total) * 100)
```

# PROOF: Mean vs Percentile using Outlier

```python
# adding artificial outlier
# added fifty to the ned of  the sepal length array
sepal_outlier = np.append(sepal_length, 50)
```

```python
# comparing mean
print("Original mean:", round(np.mean(sepal_length), 2))
print('\n')
print("mean :", np.mean(sepal_outlier))
print('\n')
print("Mean with outlier:", round(np.mean(sepal_outlier), 2))
```

```python
# comparingg median /// 50th percentile
print("Original median:", np.percentile(sepal_length, 50))
print('\n')
print("Median with outlier:", np.percentile(sepal_outlier, 50))
```
